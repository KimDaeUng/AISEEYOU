```python
def select_noun_va(x):
    new_a = []
    for j in tqdm(x):
        aa = []
        for i in j:        
            if 'NNG' in i:
                aa.append(i)
        new_a.append(aa)
    return new_a
```

```python
def only_words(x):
    a = []
    for i in tqdm(x):
        b = []
        for j in i:
            b.append(j.split('/')[0])
        a.append(b)
    return a
```

```python
def del_none(x):
    d = []
    for j in tqdm(x):
        c = []
        for i in j:
            if i == None:
                pass
            else:
                c.append(i)
        d.append(c)
    return d
```

```python
def Jaccard_similarity(doc1, doc2):
    doc1 = set(doc1)
    doc2 = set(doc2)
    return len(doc1 & doc2) / len(doc1 | doc2)
```
