```python
import pandas as pd
```

```python
df = pd.read_csv('poetry_real_final.csv')
df.head()
```

```python
import re

def to_hangul(text):
    text = re.compile('[^ 가-힣|0-9]').sub('',text) 
    return text
```

```python
def remove_bracket(row):
    pattern = r'\[.*?\]'
    return re.sub(pattern, '', row)
```

```python
df['title'] = df['title'].apply(to_hangul)
df['author'] = df['author'].apply(to_hangul)
```

```python
author = ''
for idx, row in df.iterrows():
    if author != row['author']:
        author = row['author']
    
    poem_txt = []
    
    title = row['title']
    if author in title:
        title = title.replace(author, '')
    if ('오애숙' in author) and '은파' in title:
        title = title.replace('은파', '')
    title=title.strip()
    
    tmp = remove_bracket(str(row['contents']))
    poem_txt.append(tmp)
    
    try:
        with open("./txt/{author}-{title}.txt".format(author=author, title=title), "w") as txt_file:
            for t in poem_txt:
                if type(t) != type('p'):
                    continue
                tmp = t.split('+++')
                for ta in tmp:
                    txt_file.write("".join(ta) + "\n")
    except:
        print(idx, author, title)
```

## < CHI > 이런 거 제거

```python
from copy import copy 
df2 = copy(df) 
```

```python
def remove_specific(row):
    pattern = r'\<.*?\>'
    return re.sub(pattern, '', row)
```

```python
df2['contents'] = df2['contents'].apply(remove_specific)
df2['contents'] = df2['contents'].apply(remove_bracket)
```

```python
author = ''
for idx, row in df2.iterrows():
    if author != row['author']:
        author = row['author']
        author = author.replace(' ', '')
    
    poem_txt = []
    
    title = row['title']
    if author in title:
        title = title.replace(author, '')
    if ('오애숙' in author) and '은파' in title:
        title = title.replace('은파', '')
        
    title=title.strip()
    
    poem_txt.append(str(row['contents']))
    
    try:
        with open("./txt-removed/{author}-{title}.txt".format(author=author, title=title), "w") as txt_file:
            for t in poem_txt:
                if type(t) != type('p'):
                    continue
                tmp = t.split('+++')
                for ta in tmp:
                    txt_file.write("".join(ta) + "\n")
    except:
        print(idx, author, title)
```
