```python
import pandas as pd
import numpy as np
```

```python
poetry = pd.read_csv('poetry_real_final.csv')
poetry
```

```python
author_list = ['김소월','김수영1', '김춘수', '나희덕', '노천명', '노혜경', '류시화', '복효근', '서정주', '유치환', '윤인환', '정지용', '천상병', '한용운', '고은-', '김낙필', '도종환', '박목월', '서지월','안도현', '용혜원', '정호승', '황인숙']
```

```python
poetry2 = pd.DataFrame()
```

```python
for author in author_list:
    d = poetry[poetry.author == author]
    poetry2 = pd.concat([poetry2,d],axis=0)
```

```python
poetry2 = poetry2.loc[:,['title','author','contents']].reset_index(drop = True)
```

```python
poetry2.to_csv('poetry_summary.csv',index=False)
```
