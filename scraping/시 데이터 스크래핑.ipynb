```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 수집해야 할 데이터 url
# 총 3611 페이지

page_urls = []
for i in range(3611):
    page_urls.append('http://www.poemlove.co.kr/bbs/board.php?bo_table=tb01&page=1&page={}'.format(i+1))
```

```python
wa = []
aa = []
ta = []

for i in range(len(page_urls)):
    req = requests.get(page_urls[i])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    writer = soup.select('td > nobr')[2:]
    for w in writer:
        wa.append(w.text)

    author = soup.select("td > a > font")[24:]
    for a in author:
        aa.append(a.text)

    title = soup.select("tr > td > nobr > a > span")[2:]
    for t in title:
        ta.append(t.text)
```
