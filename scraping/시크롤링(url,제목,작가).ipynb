```python
urls = []
titles = []
name = []
for url in tqdm_notebook(page_urls):
    req = requests.get(url)
    root = lxml.html.fromstring(req.text)
    for i in root.cssselect('li a')[16:]:
        urls.append(i.attrib['href'])
    for j in root.cssselect('span.wsubject')[2:]:
        titles.append(j.text_content().strip())
    for k in range(3,len(root.cssselect('span.wname')[3:])+2):
        if k % 2==0:
            name.append(re.compile('[가-힣]+').findall(root.cssselect('span.wname')[k].text_content()))
```
